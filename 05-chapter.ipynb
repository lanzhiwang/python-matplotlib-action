### 刻度定位器和刻度格式器的使用方法

```python
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter

x = np.linspace(-0.1, 1, 20)
y = x

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)

# 主刻度线
ax.xaxis.set_major_locator(MultipleLocator(1.0))
ax.yaxis.set_major_locator(MultipleLocator(0.2))

# 次要刻度线
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))


def minor_tick(x, pos):
    if not x % 1.0:
        return ""
    return "%.2f" % x


# 刻度值显示精度
ax.xaxis.set_minor_formatter(FuncFormatter(minor_tick))

# 刻度线样式
ax.tick_params("y", which="major", length=5, width=2.0, colors="r")
ax.tick_params("x", which="minor", length=5, width=1.0, labelsize=10, colors="0.25")

ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-0.1, 1.1)

ax.plot(x, y, c=(0.25, 0.25, 1.00), lw=2, zorder=10)

plt.show()
```

### 有指示注解和无指示注解的添加方法

```python
x = np.linspace(-0.05, 2 * np.pi, 50)
y = np.sin(x)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)

ax.plot(x, y, c="b", ls="--", lw=2)

ax.annotate(
    "maximum",
    xy=(np.pi / 2, 1.0),
    xycoords="data",  # xy 坐标系, xycoords 参数值为 data 表示 xy 坐标与折线图使用相同的坐标系
    xytext=((np.pi / 2) + 0.15, 0.8),
    weight="bold",
    color="r",
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="r"),
)

ax.text(
    2.8,
    0.4,
    r"$y=\sin(x)$",
    fontsize=20,
    color="b",
    bbox=dict(facecolor="y", alpha=0.5),
)
plt.show()
```

### 案例1: 圆角文本框的设置

```python
x = np.linspace(-0.05, 2 * np.pi, 50)
y = np.sin(x)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)

ax.plot(x, y, c="b", ls="--", lw=2)

ax.annotate(
    "maximum",
    xy=(np.pi / 2, 1.0),
    xycoords="data",  # xy 坐标系, xycoords 参数值为 data 表示 xy 坐标与折线图使用相同的坐标系
    xytext=((np.pi / 2) + 0.15, 0.8),
    weight="bold",
    color="r",
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="r"),
)

ax.text(
    2.8,
    0.4,
    r"$y=\sin(x)$",
    fontsize=20,
    color="b",
    bbox=dict(
        boxstyle="round", facecolor="y", alpha=0.5
    ),  # 圆角文本框的设置 boxstyle="round"
)
plt.show()
```

### 案例3: 圆角线框的有弧度指示的注解

```python
x = np.linspace(-0.05, 2 * np.pi, 50)
y = np.sin(x)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)

ax.plot(x, y, c="b", ls="--", lw=2)

ax.annotate(
    "maximum",
    xy=(np.pi / 2, 1.0),
    xycoords="data",  # xy 坐标系, xycoords 参数值为 data 表示 xy 坐标与折线图使用相同的坐标系
    xytext=((np.pi / 2) + 0.15, 0.8),
    weight="bold",
    color="r",
    bbox=dict(boxstyle="round", facecolor="y", alpha=0.5),
    arrowprops=dict(
        arrowstyle="-|>", connectionstyle="angle", color="r"
    ),  ### 圆角线框的有弧度指示的注解 arrowstyle="-|>", connectionstyle="angle"
)

ax.text(
    2.8, 0.4, r"$y=\sin(x)$", fontsize=20, color="b", alpha=0.5
)  # 文本的水印效果 alpha
plt.show()
```

### 案例4: 有箭头指示的趋势线

```python
x = np.linspace(-0.05, 2 * np.pi, 50)
y = np.sin(x)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)

ax.plot(x, y, c="b", ls="--", lw=2)

ax.annotate(
    "",
    xy=(3 * np.pi / 2, np.sin(3 * np.pi / 2) + 0.05),
    xycoords="data",  # xy 坐标系, xycoords 参数值为 data 表示 xy 坐标与折线图使用相同的坐标系
    xytext=(np.pi / 2, np.sin(np.pi / 2) + 0.05),
    color="r",
    arrowprops=dict(arrowstyle="->", color="r"),
)

# arrow(x, y, dx, dy)
ax.arrow(0.0, -0.4, np.pi / 2, 1.2, head_width=0.05, head_length=0.1, fc="g", ec="g")


plt.show()
```

### 案例5: 桑基图

```python
from matplotlib.sankey import Sankey

"""
列表 flows 中的负值表示流出量, 正值表示流入量.
列表 orientations 中的 -1、0 和 1 分别表示流量的显示位置在下方、水平和上方.
最后, 调整流量图 diagrams[0] 的文本 "List Cost" 和 "family" 的显示样式、颜色等属性的属性值.
"""
flows = [0.2, 0.1, 0.4, 0.3, -0.6, -0.05, -0.15, -0.2]
labels = ["c1", "c2", "c3", "c4", "c5", "c6", "c7", "c8"]
orientations = [1, 1, 0, -1, 1, -1, 1, 0]

sankey = Sankey()
sankey.add(
    flows=flows,
    labels=labels,
    orientations=orientations,
    color="c",
    fc="lightgreen",
    patchlabel="Life Cost",
    alpha=0.7,
)

diagrams = sankey.finish()
diagrams[0].texts[4].set_color("r")
diagrams[0].texts[4].set_weight("bold")
diagrams[0].text.set_fontsize(20)
diagrams[0].text.set_fontweight("bold")

plt.show()
```

### 实现标题和坐标轴标签的投影效果的操作方法

```python
import matplotlib.patheffects as pes

x = np.linspace(-0.05, 2 * np.pi, 50)
y = np.sin(x)

plt.plot(x, y, c="b", ls="--", lw=2)

title = plt.title(r"$y = \sin(x)$", fontsize=23, va="bottom")
xlabel = plt.xlabel("xlabel", fontsize=20, alpha=1.0)
ylabel = plt.ylabel("ylabel", fontsize=20, alpha=1.0)


title.set_path_effects([pes.withSimplePatchShadow()])
pe = pes.withSimplePatchShadow(offset=(1, -1), shadow_rgbFace="r", alpha=0.8)
xlabel.set_path_effects([pe])
ylabel.set_path_effects([pe])

plt.show()
```

### 案例1: 给坐标轴标签添加文本框

```python
x = np.linspace(-0.05, 2 * np.pi, 50)
y = np.sin(x)

plt.plot(x, y, c="b", ls="--", lw=2)

plt.title(r"$y = \sin(x)$", fontsize=23, va="bottom")

box = dict(facecolor="#6959CD", pad=2, alpha=0.4)
plt.xlabel("xlabel", fontsize=20, alpha=1.0, bbox=box)
plt.ylabel("ylabel", fontsize=20, alpha=1.0, bbox=box)

plt.gca().yaxis.set_label_coords(-0.08, 0.2)  # axes coords
plt.gca().xaxis.set_label_coords(1.0, -0.05)  # axes coords

plt.show()
```
