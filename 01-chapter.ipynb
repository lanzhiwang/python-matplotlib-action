### List of named colors

```python
import math

import matplotlib.pyplot as plt

import matplotlib.colors as mcolors
from matplotlib.patches import Rectangle


def plot_colortable(colors, *, ncols=4, sort_colors=True):

    cell_width = 212
    cell_height = 22
    swatch_width = 48
    margin = 12

    # Sort colors by hue, saturation, value and name.
    if sort_colors is True:
        names = sorted(
            colors, key=lambda c: tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(c)))
        )
    else:
        names = list(colors)

    n = len(names)
    nrows = math.ceil(n / ncols)

    width = cell_width * ncols + 2 * margin
    height = cell_height * nrows + 2 * margin
    dpi = 72

    fig, ax = plt.subplots(figsize=(width / dpi, height / dpi), dpi=dpi)
    fig.subplots_adjust(
        margin / width,
        margin / height,
        (width - margin) / width,
        (height - margin) / height,
    )
    ax.set_xlim(0, cell_width * ncols)
    ax.set_ylim(cell_height * (nrows - 0.5), -cell_height / 2.0)
    ax.yaxis.set_visible(False)
    ax.xaxis.set_visible(False)
    ax.set_axis_off()

    for i, name in enumerate(names):
        row = i % nrows
        col = i // nrows
        y = row * cell_height

        swatch_start_x = cell_width * col
        text_pos_x = cell_width * col + swatch_width + 7

        ax.text(
            text_pos_x,
            y,
            name,
            fontsize=14,
            horizontalalignment="left",
            verticalalignment="center",
        )

        ax.add_patch(
            Rectangle(
                xy=(swatch_start_x, y - 9),
                width=swatch_width,
                height=18,
                facecolor=colors[name],
                edgecolor="0.7",
            )
        )

    return fig
```

```python
plot_colortable(mcolors.BASE_COLORS, ncols=3, sort_colors=False)
```

```python
plot_colortable(mcolors.TABLEAU_COLORS, ncols=2, sort_colors=False)
```

```python
plot_colortable(mcolors.CSS4_COLORS)
plt.show()
```

## 函数 xlabel() ylabel() - 设置 x 轴和 y 轴的标签文本

```python
x = np.linspace(1, 10, 10)
y = np.random.randn(10)

plt.scatter(x, y, label="scatter figure", marker="*")
plt.xlim(1, 10)
plt.ylim(1, 10)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

### 设置 xlabel ylabel 参数

```python
# https://matplotlib.org/3.10.0/api/text_api.html

# Valid font size are xx-small, x-small, small, medium, large, x-large, xx-large, larger, smaller, None
# Vertical alignment must be one of ("top", "bottom", "center", "baseline")
# Horizontal alignment must be one of ("center", "right", "left")
fontdict = {
    "fontsize": "xx-large",
    "verticalalignment": "baseline",
    "horizontalalignment": "center",
}

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=3, label="plot figure")
plt.xlabel("x-axis", fontdict=fontdict)
plt.ylabel("y-axis")
plt.legend()  # 显示 label="plot figure" 文本
plt.show()
```

## 函数 axhline() 一 绘制平行于 x 轴的水平参考线
## 函数 axvline() 一 绘制平行于 y 轴的竖直参考线

Signature: plt.axhline(y=0, xmin=0, xmax=1, hold=None, **kwargs)

- y 水平参考线的出发点
- c 参考线的线条颜色
- ls 参考线的线条风格
- lw 参考线的线条宽度


```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=3, label="plot figure")

plt.axhline(y=0, xmin=0, xmax=1, c="r", ls="--", lw=2)

plt.legend()  # 显示 label="plot figure" 文本
plt.show()
```

```python
"""
xmin : scalar, optional, default: 0
    Should be between 0 and 1, 0 being the far left of the plot, 1 the far right of the plot.
xmax : scalar, optional, default: 1
    Should be between 0 and 1, 0 being the far left of the plot, 1 the far right of the plot.
"""

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=3, label="plot figure")

plt.axhline(y=0, xmin=0.2, xmax=0.4, c="r", ls="--", lw=2)

plt.legend()  # 显示 label="plot figure" 文本
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=3, label="plot figure")

plt.axhline(y=0, xmin=0.2, xmax=0.4, c="r", ls="--", lw=2)
plt.axvline(4, ymin=0.6, ymax=0.8, color="g", linestyle=":", linewidth=2)
plt.legend()  # 显示 label="plot figure" 文本
plt.show()
```

## 函数 axvspan() 一 绘制垂直于 x 轴的参考区域
## 函数 axhspan() 一 绘制垂直于 y 轴的参考区域

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=3, label="plot figure")

plt.axvspan(
    xmin=4,  # 起始位置
    xmax=6,  # 终止位置
    ymin=0.4,  # 起始比例
    ymax=1,
    facecolor="y",  # 填充颜色
    alpha=0.3,  # 填充透明度
)
plt.legend()  # 显示 label="plot figure" 文本
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=3, label="plot figure")

plt.axvspan(
    xmin=4,  # 起始位置
    xmax=6,  # 终止位置
    ymin=0.4,  # 起始比例
    ymax=1,
    facecolor="y",  # 填充颜色
    alpha=0.3,  # 填充透明度
)
plt.axhspan(ymin=0, ymax=0.6, xmin=0, xmax=1, facecolor="b", alpha=0.3)
plt.legend()  # 显示 label="plot figure" 文本
plt.show()
```

## 函数 annotate() — 添加图形内容细节的指向型注释文本

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, label="plot figure")

plt.annotate(
    "maximum",  # 注释文本
    xy=(np.pi / 2, 1.0),  # 被注释图像内容的位置坐标(箭头指向的地方, 即需要标注的坐标点)
    xytext=(
        (np.pi / 2) + 1.0,
        0.8,
    ),  # 注释文本的位置坐标(箭头开始指的地方, 即放置注释文本的地方)
    color="b",  # 注释文本的颜色
    arrowprops=dict(
        arrowstyle="->", connectionstyle="arc3", color="b"
    ),  # 箭头的属性字典
)
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, label="plot figure")

plt.annotate(
    "maximum",  # 注释文本
    xy=(
        np.pi / 2,
        1.0,
    ),  # 被注释图像内容的位置坐标 (箭头指向的地方, 即需要标注的坐标点)
    xytext=(
        (np.pi / 2) + 1.0,
        0.8,
    ),  # 注释文本的位置坐标(箭头开始指的地方, 即放置注释文本的地方)
    color="b",  # 注释文本的颜色
    arrowprops=dict(
        arrowstyle="-|>", connectionstyle="angle", color="r"
    ),  # 箭头的属性字典
)

plt.annotate(
    "minimum",  # 注释文本
    xy=(
        np.pi * 1.5,
        -1.0,
    ),  # 被注释图像内容的位置坐标 (箭头指向的地方, 即需要标注的坐标点)
    xytext=(
        np.pi * 1.25,
        -0.5,
    ),  # 注释文本的位置坐标(箭头开始指的地方, 即放置注释文本的地方)
    color="b",  # 注释文本的颜色
    arrowprops=dict(
        arrowstyle="-|>", connectionstyle="angle3", color="r"
    ),  # 箭头的属性字典
)
plt.legend()
plt.show()
```

## 函数 text() — 添加图形内容细节的无指向型注释文本

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, label="plot figure")
plt.text(x=np.pi + 0.5, y=0.0, s="y=sin(x)", weight="bold", color="b", fontsize=12)
plt.legend()
plt.show()
```

## 函数 title() — 添加图形内容的标题

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, label="plot figure")
plt.title("y=sin(x)")
plt.legend()
plt.show()
```

## 函数 legend() — 标示不同图形的文本标签图例

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, label="plot figure")
plt.title("y=sin(x)")
plt.legend(loc="lower right")
plt.show()
```

## 综合示例

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.random.randn(100)

plt.plot(x, y1, ls="-.", lw=2, label="plot figure")

plt.scatter(x, y2, label="scatter figure", marker="*")

## xlim
plt.xlim(0, 8)
plt.ylim(-3, 3)

## xlabel
plt.xlabel("x-axis")
plt.ylabel("y-axis")

## grid 刻度线
plt.grid(color="r", linestyle=":")

## axhline 水平参考线
plt.axhline(y=0.5, color="r", xmin=0.5, xmax=1, linestyle="--", linewidth=2)

plt.axvline(4, ymin=0.6, ymax=0.8, color="g", linestyle=":", linewidth=2)

## axvspan 竖直参考区域
plt.axvspan(
    xmin=1,  # 起始位置
    xmax=3,  # 终止位置
    ymin=0,
    ymax=1,
    facecolor="y",  # 填充颜色
    alpha=0.3,  # 填充透明度
)
plt.axhspan(ymin=0, ymax=1, xmin=0, xmax=1, facecolor="g", alpha=0.3)

## annotate 指向型文本注释
plt.annotate(
    "maximum",  # 注释文本
    xy=(
        np.pi / 2,
        1.0,
    ),  # 被注释图像内容的位置坐标 (箭头指向的地方, 即需要标注的坐标点)
    xytext=(
        (np.pi / 2) + 1.0,
        0.8,
    ),  # 注释文本的位置坐标(箭头开始指的地方, 即放置注释文本的地方)
    color="b",  # 注释文本的颜色
    arrowprops=dict(
        arrowstyle="->", connectionstyle="arc3", color="b"
    ),  # 箭头的属性字典
)

## text 无指向性文本注释

plt.title("y=sin(x)")
plt.text(x=np.pi, y=0.0, s="y=sin(x)", weight="bold", color="b")

#################################################
# clean up (removing chart junk)
# turn off the top spine and the right spine
# 关闭上面和右边的坐标轴
for spine in plt.gca().spines.keys():
    if spine == "top" or spine == "right":
        plt.gca().spines[spine].set_color("none")

# 下面这两句可以不写, 默认就打开了
# turn on the bottom tick for x-axis
plt.gca().xaxis.set_ticks_position("bottom")
# set tick_line position of the bottom

# turn on the left ticks for y-axis
plt.gca().yaxis.set_ticks_position("left")
# set tick_line position of the left

#################################################

# plt.legend(loc="lower right")
plt.show()
```
