## 折线图

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

# ls="-." 折线图的线条风格
# lw=3 折线图的线宽
plt.plot(x, y, ls="-.", lw=3, label="plot figure")
plt.legend()  # 显示 label="plot figure" 文本
plt.show()
```

## 散点图

```python
x = np.linspace(1, 10, 10)
y = np.random.randn(10)

plt.scatter(x, y, c="b", label="scatter figure")
plt.legend()
plt.show()
```

## 柱状图

```python
x = list(range(1, 11))
y = [3, 1, 4, 5, 6, 8, 9, 7, 2, 10]

plt.bar(x, y, align="center", color="c", tick_label=list("abcdefghij"), hatch="/")
plt.show()
```

```python
x = list(range(1, 11))
y = [3, 1, 4, 5, 6, 8, 9, 7, 2, 10]

plt.bar(
    x,
    y,
    width=0.2,
    align="center",
    color="c",
    tick_label=list("abcdefghij"),
    hatch="\\",
)
plt.show()
```

```python
x = list(range(1, 11))
y = [3, 1, 4, 5, 6, 8, 9, 7, 2, 10]

plt.bar(
    x,
    y,
    width=0.8,
    align="center",
    color="c",
    tick_label=list("abcdefghij"),
    hatch="\\",
    edgecolor="r",
)
plt.show()
```

## 水平柱状图/条形图

```python
x = list(range(1, 11))
y = [3, 1, 4, 5, 6, 8, 9, 7, 2, 10]

plt.barh(
    x,
    y,
    align="center",
    color="c",
    tick_label=list("abcdefghij"),
    hatch="\\",
    edgecolor="r",
)
plt.show()
```

## 直方图

直方图的高度表示落在该区间内的元素的个数

```python
box_hight = np.random.randint(low=0, high=10, size=100)
print("box_hight:", box_hight)
bins = range(0, 11)
print("bins:", list(bins))
plt.hist(box_hight, bins)
plt.show()
```

```python
box_hight = np.random.randint(low=0, high=10, size=100)
print("box_hight:", box_hight)
bins = range(0, 11)
print("bins:", list(bins))
plt.hist(x=box_hight, bins=bins, histtype="bar", rwidth=1, alpha=0.6)
plt.show()
```

```python
box_hight = np.random.randint(low=0, high=10, size=100)
print("box_hight:", box_hight)
bins = range(0, 11)
print("bins:", list(bins))
plt.hist(
    x=box_hight, bins=bins, histtype="bar", rwidth=1, alpha=0.6, cumulative=True
)  # 累加
plt.show()
```

## 饼图

```python
sold_nums = [0.05, 0.45, 0.15, 0.35]
kinds = "class1", "class2", "class3", "class4"
colors = ["#e41a1c", "#377eb8", "#4daf4a", "#984ea3"]
plt.pie(x=sold_nums, labels=kinds, autopct="%3.1f%%", startangle=0, colors=colors)
plt.show()
```

```python
sold_nums = [0.05, 0.45, 0.15, 0.35]
kinds = "class1", "class2", "class3", "class4"
colors = ["#e41a1c", "#377eb8", "#4daf4a", "#984ea3"]
plt.pie(
    x=sold_nums,
    labels=kinds,
    autopct="%3.1f%%",
    startangle=60,
    colors=colors,
    shadow=True,  # 阴影
)
plt.show()
```

## 极线图

```python
bar_slices = 12
theta = np.linspace(start=0.02, stop=2 * np.pi, num=bar_slices, endpoint=False)
print("theta:", theta)
r = 30 * np.random.rand(bar_slices)
print("r:", r)
plt.polar(
    theta, r, linewidth=2, marker="*", mfc="b", ms=10, runits="km", thetaunits="rad"
)
plt.show()
```

```python
bar_slices = 12
theta = np.linspace(start=0.02, stop=2 * np.pi, num=bar_slices, endpoint=False)
print("theta:", theta)
r = 30 * np.random.rand(bar_slices)
print("r:", r)
plt.polar(
    theta, r, linewidth=2, marker="*", mfc="b", ms=30, runits="km", thetaunits="rad"
)
plt.show()
```

## 气泡图

```python
# 函数 randn(8) 表示在标准正态分布中随机地取 8 个数
x = np.random.randn(8)
y = np.random.randn(8)

plt.scatter(
    x,
    y,
    s=np.power(x + 10 * y, 2),  # 标记点的大小
    marker="o",
    c=np.random.rand(8),  # 颜色
)
plt.show()
```

## 棉签图

```python
x = np.linspace(0.5, 2 * np.pi, 20)
print("x:", x)
y = np.random.randn(20)
print("y:", y)

plt.stem(
    x,
    y,
    linefmt="-.",  # 棉棒
    markerfmt="o",  # 棉签头
    basefmt="-",  # 水平基线
    label="stem fig",
)
plt.legend()
plt.show()
```

## 箱线图

```python
x = np.random.randn(10)
print("x:", x)
plt.boxplot(x)
plt.show()
```

## 误差棒图

```python
x = np.linspace(0.1, 0.6, 6)
print("x:", x)
y = np.exp(x)
print("y:", y)
plt.errorbar(x, y, fmt="ro", yerr=0.2, xerr=0.02, ecolor="b")
plt.xlim(0, 0.7)
plt.show()
```
