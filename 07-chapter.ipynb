### 共享单一绘图区域的坐标轴

`ax.twinx()` or `ax.twiny()`

```python
t = np.arange(0.05, 10.0, 0.01)
s1 = np.exp(t)
s2 = np.cos(t)

fig, ax = plt.subplots()  # 默认一行一列
ax.plot(t, s1, c="b", ls="--", label="$\exp(x)$")
ax.set_xlabel("x-axis")  # 设置 x 坐标轴
ax.set_ylabel("exp(x)", color="b")  # 设置 y 坐标轴
ax.tick_params("y", colors="b")  # y 坐标轴的刻度
ax.legend()

# 关键, 共享 x 坐标轴
ax2 = ax.twinx()  # Create a twin Axes sharing the xaxis
ax2.plot(t, s2, c="g", ls=":", label="$\cos(x)$")
ax2.set_ylabel("cos", color="r")  #  设置 y 坐标轴
ax2.tick_params("y", colors="r")
ax2.legend()

plt.title("twinx")
plt.show()
```

```python
t = np.arange(0.05, 10.0, 0.01)
s1 = np.exp(t)
s2 = np.cos(t)

fig, ax = plt.subplots()  # 默认一行一列
ax.plot(t, s1, c="b", ls="--", label="$\exp(x)$")
ax.set_xlabel("x-axis")  # 设置 x 坐标轴
ax.set_ylabel("exp(x)", color="b")  # 设置 y 坐标轴
ax.tick_params("y", colors="b")  # y 坐标轴的刻度
ax.legend()

# 关键, 共享 x 坐标轴
ax2 = ax.twinx()  # Create a twin Axes sharing the xaxis
ax2.plot(t, s2, c="g", ls=":", label="$\cos(x)$")
ax2.set_ylabel("cos", color="r")  #  设置 y 坐标轴
ax2.tick_params(
    direction="out", length=6, width=2, colors="r", grid_color="k", grid_alpha=0.5
)
ax2.legend()

plt.title("twinx")
plt.show()
```

### 共享不同子区绘图区域的坐标轴

调整 `subplots()` 中的 `sharex` 或 `sharey` 参数

sharex 与 sharey 的取值共有 4 中, `row`、`col`、`all`、`None`, 其中 all 和 none 分别等同于 true 和 false

```python
x1 = np.linspace(-10, 2 * np.pi, 100)
y1 = np.sin(x1)

x2 = np.linspace(0, 4, 100)
y2 = np.random.randn(100)

x3 = np.linspace(0, 2 * np.pi, 100)
y3 = np.cos(x3**2)

x4 = np.linspace(0, 4, 10)
y4 = np.power(x4, 2)

fig, ax = plt.subplots(2, 2, sharex=None, sharey=None)  # x 轴和 y 轴均不共享

ax[0, 0].plot(x1, y1)
ax[0, 1].scatter(x2, y2)
ax[1, 0].plot(x3, y3)
ax[1, 1].scatter(x4, y4)

plt.show()
```

```python
x1 = np.linspace(-10, 2 * np.pi, 100)
y1 = np.sin(x1)

x2 = np.linspace(0, 4, 100)
y2 = np.random.randn(100)

x3 = np.linspace(0, 2 * np.pi, 100)
y3 = np.cos(x3**2)

x4 = np.linspace(0, 4, 10)
y4 = np.power(x4, 2)

fig, ax = plt.subplots(2, 2, sharex="all", sharey=None)  # x 轴共享, y 轴均不共享

ax[0, 0].plot(x1, y1)
ax[0, 1].scatter(x2, y2)
ax[1, 0].plot(x3, y3)
ax[1, 1].scatter(x4, y4)

plt.show()
```

### 共享个别子区绘图区域的坐标轴

```python
x1 = np.linspace(-10, 2 * np.pi, 100)
y1 = np.sin(x1)

x2 = np.linspace(0, 4, 100)
y2 = np.random.randn(100)

x3 = np.linspace(0, 2 * np.pi, 100)
y3 = np.cos(x3**2)

x4 = np.linspace(0, 4, 10)
y4 = np.power(x4, 2)

# fig, ax = plt.subplots(2, 2)
ax1 = plt.subplot(221)
ax2 = plt.subplot(222)
ax3 = plt.subplot(223, sharey=ax2)  # 2 与 3 共享 y 轴
ax4 = plt.subplot(224, sharex=ax1)  # 1 与 4 共享 x 轴

ax1.plot(x1, y1)
ax2.scatter(x2, y2)
ax3.plot(x3, y3)
ax4.scatter(x4, y4)

plt.show()
```
