## 阶梯图

````python
x = np.linspace(0, 10, 40)
y = np.sin(x)
plt.plot(x, y, label="y=sin(x)")
"""where: [ "pre" | "post" | "mid" ]
If "pre" (the default), the interval from x[i] to x[i+1] has level y[i+1].
If "post", that interval has level y[i].
If "mid", the jumps in y occur half-way between the x-values.
"""
plt.step(x, y, where="pre", label="y=step(sin(x))-pre")

plt.legend()
plt.show()
````

````python
x = np.linspace(0, 10, 40)
y = np.sin(x)
plt.plot(x, y, label="y=sin(x)")
"""where: [ "pre" | "post" | "mid" ]
If "pre" (the default), the interval from x[i] to x[i+1] has level y[i+1].
If "post", that interval has level y[i].
If "mid", the jumps in y occur half-way between the x-values.
"""
plt.step(x, y, where="mid", label="y=step(sin(x))-mid")

plt.legend()
plt.show()
````

## 误差棒图

````python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
error = 0.05 + 0.15 * x

plt.plot(x, y, ls="-.", lw=3)

lower_error = error
upper_error = error * 0.15

error_limit = [lower_error, upper_error]

plt.errorbar(
    x,
    y,
    yerr=error_limit,  # 单一数值的非对称形式误差范围
    fmt="o",  # 数据点的标记样式和连接线样式
    ecolor="y",  # 误差棒的线条颜色
    elinewidth=4,  # 误差棒的线宽
    ms=5,  # 数据点的大小
    mfc="c",  # 数据点的颜色
    mec="r",  # 数据点的边缘颜色
    capsize=2,
    capthick=1,  # 误差棒边界横杠大小
)

plt.show()
````

### 带误差棒的柱状图

````python
x = np.arange(5)
y = [100, 82, 39, 92, 43]
std_err = [7, 2, 6, 10, 5]

err_attribute = dict(elinewidth=2, ecolor="black", capsize=3)
plt.bar(
    x,
    y,
    color="c",
    width=0.6,
    align="center",
    yerr=std_err,  # 关键——指定误差值
    error_kw=err_attribute,  # 误差bar的绘制属性
    tick_label=list("ABCDE"),
)

plt.show()
````

### 带误差棒的条形图

````python
x = list(range(1, 11))
y = [3, 1, 4, 5, 6, 8, 9, 7, 2, 10]
std_err = [7, 2, 6, 10, 5, 1, 2, 3, 4, 5]

err_attribute = dict(elinewidth=2, ecolor="black", capsize=3)
plt.barh(
    x,
    y,
    align="center",
    color="c",
    tick_label=list("abcdefghij"),
    hatch="\\",
    edgecolor="r",
    xerr=std_err,
    error_kw=err_attribute,
)
plt.show()
````

### 带误差棒的多数据并列柱状图

````python
x = np.arange(5)
y1 = [100, 82, 39, 92, 43]
y2 = [83, 92, 100, 53, 25]
std_err1 = [7, 2, 6, 10, 5]
std_err2 = [6, 5, 10, 2, 7]

bar_width = 0.35
err_attribute = dict(elinewidth=2, ecolor="black", capsize=3)

# 第一组
plt.bar(
    x,
    y1,
    color="c",
    width=bar_width,
    align="center",
    yerr=std_err1,
    error_kw=err_attribute,
    label="group 1",
)
# 第二组
plt.bar(
    x + bar_width,  # 并列摆放
    y2,
    color="m",
    width=bar_width,
    align="center",
    yerr=std_err2,
    error_kw=err_attribute,
    label="group 2",
)

plt.grid(axis="y", linestyle=":", color="gray", alpha=0.4)
plt.show()
````
