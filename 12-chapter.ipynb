### 颜色映射表的使用

#### 获取全部可用的颜色映射表

- plt.colormaps()

#### 注册新的颜色映射表到 matplotlib 中

- matplotlib.cm.register_cmap()


```python
plt.colormaps()
```

```python
random_data = np.random.rand(20, 20)
plt.pcolor(random_data, cmap="BuPu")  # Create a pseudocolor plot of a 2-D array.
plt.colorbar()
plt.show()
```

```python
random_data = np.random.rand(20, 20)
plt.pcolor(random_data, cmap=matplotlib.cm.BuPu)  # 指定颜色格式colormap的另一种方法
plt.colorbar()
plt.show()
```
