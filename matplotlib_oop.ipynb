## matplotlib 面向对象

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig = plt.figure()
print("fig:", fig)

axes = fig.get_axes()
print("axes:", axes)

axes = fig.add_axes((0.1, 0.1, 1, 1))
print("axes:", axes)

# lines = axes.plot(x, y, ls="-.", lw=3, label="plot figure")
lines = axes.plot(x, y)
for line in lines:
    line.set_linewidth(3)
    line.set_linestyle("--")
    line.set_label("plot figure")

legend = axes.legend()
legend.set_loc("upper left")

fig.show()
```
